```python
class LogitRegression():
    
    def __init__(self, n, learning_rate):
        self.n = n # number of iterations
        self.learning_rate = learning_rate
    
    def sigmoid(self, X, W):
        
        # p = np.dot(X, W)
        Y_predict = 1/(1 + np.exp(-(np.dot(X, W).astype(np.float))))
        return Y_predict
    
    def cost(self, X, W, Y):
        Y_predict = self.sigmoid(X, W)
        cost = np.sum(-(Y*np.log(Y_predict))-((1 - Y)*(np.log(1 - Y_predict))))
        cost = cost/len(Y)
        return cost
    
    def gradient_descent(self, X, W, Y):
       
        Y_predict = self.sigmoid(X, W)
        delta = Y_predict - Y
        dW = (np.dot(X.T,delta))/len(Y)
        W = W  - self.learning_rate*dW
        return W
            
    def fit(self, X, W, Y):
        loss = []
        for i in range(self.n):
            W = self.gradient_descent(X, W, Y)
            loss.append(self.cost(X, W, Y))
        return W, loss
        
    def predict(self, X, W):
        Y_predict = self.sigmoid(X, W)
        Y_predicted = np.where(Y_predict >= 0.5, 1, 0)
        return Y_predicted
```
